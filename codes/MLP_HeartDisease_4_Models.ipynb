# چهار ترکیب مختلف برای مدل MLP
این نوت‌بوک شامل ۴ ترکیب مختلف از مدل MLP بر اساس جدول پیشنهادی است.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD 
import matplotlib.pyplot as plt
```

```python
# بارگذاری داده‌ها
df = pd.read_csv('Heart_disease_cleveland_new.csv')
X = df.drop('target', axis=1).values
y = df['target'].values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_val, y_train, y_val = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# ترکیب 1
model_1 = Sequential()
model_1.add(Dense(32, input_dim=X_train.shape[1], activation='relu'))
model_1.add(Dense(1, activation='sigmoid'))
model_1.compile(optimizer=SGD(learning_rate=0.01), loss='binary_crossentropy', metrics=['accuracy'])
history_1 = model_1.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=30, batch_size=32, verbose=1)
```

```python
# ترکیب 2
batch_size_gd_2 = X_train.shape[0]
optimizer_gd_2 = SGD(learning_rate=0.001)

model_2 = Sequential()
model_2.add(Dense(64, input_dim=X_train.shape[1], activation='relu'))
model_2.add(Dense(32, activation='relu'))
model_2.add(Dense(1, activation='sigmoid'))
model_2.compile(optimizer=optimizer_gd_2, loss='binary_crossentropy', metrics=['accuracy'])
history_2 = model_2.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=30, batch_size=batch_size_gd_2, verbose=1)
```

```python
# ترکیب 3
batch_size_gd_3 = X_train.shape[0]
optimizer_gd_3 = SGD(learning_rate=0.01)

model_3 = Sequential()
model_3.add(Dense(16, input_dim=X_train.shape[1], activation='relu'))
model_3.add(Dense(1, activation='sigmoid'))
model_3.compile(optimizer=optimizer_gd_3, loss='binary_crossentropy', metrics=['accuracy'])
history_3 = model_3.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=30, batch_size=batch_size_gd_3, verbose=1)
```

```python
# ترکیب 4
model_4 = Sequential()
model_4.add(Dense(32, input_dim=X_train.shape[1], activation='relu'))
model_4.add(Dense(16, activation='relu'))
model_4.add(Dense(1, activation='sigmoid'))
model_4.compile(optimizer=SGD(learning_rate=0.005), loss='binary_crossentropy', metrics=['accuracy'])
history_4 = model_4.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=30, batch_size=32, verbose=1)
```

```python
# تاثیر batch_size های مختلف بر همگرایی
batch_sizes = [1, 16, 32, 64, 128, X_train.shape[0]]
plt.figure(figsize=(14, 10))

for batch_size in batch_sizes:
    model = Sequential()
    model.add(Dense(16, input_dim=X_train.shape[1], activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    optimizer = SGD(learning_rate=0.01)
    model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
    
    # آموزش مدل
    history = model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=30, batch_size=batch_size, verbose=0)
    
    # رسم منحنی همگرایی
    plt.plot(history.history['val_accuracy'], label=f'batch_size={batch_size}')

plt.title('Impact of Batch Size on Convergence Speed and Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Validation Accuracy')
plt.legend()
plt.show()
```

```python
# رسم دقت اعتبارسنجی (Validation Accuracy) برای هر ۴ مدل
plt.figure(figsize=(14, 10))

histories = [history_1, history_2, history_3, history_4]
model_labels = ["Model 1 (SGD, 32)", "Model 2 (GD, 64, 32)", "Model 3 (GD, 16)", "Model 4 (SGD, 32, 16)"]

for i, (history, label) in enumerate(zip(histories, model_labels), 1):
    plt.plot(history.history['val_accuracy'], label=label)

plt.title("Validation Accuracy for 4 Different MLP Models")
plt.xlabel("Epochs")
plt.ylabel("Validation Accuracy")
plt.legend()
plt.show()
```

```python
# رسم دقت (Accuracy) و خطا (Loss) برای هر ۴ مدل
plt.figure(figsize=(18, 12))

# Accuracy Plot
plt.subplot(2, 1, 1)
for i, (history, label) in enumerate(zip(histories, model_labels), 1):
    plt.plot(history.history['val_accuracy'], label=label)
plt.title("Validation Accuracy for 4 Different MLP Models")
plt.xlabel("Epochs")
plt.ylabel("Validation Accuracy")
plt.legend()

# Loss Plot
plt.subplot(2, 1, 2)
for i, (history, label) in enumerate(zip(histories, model_labels), 1):
    plt.plot(history.history['val_loss'], label=label)
plt.title("Validation Loss for 4 Different MLP Models")
plt.xlabel("Epochs")
plt.ylabel("Validation Loss")
plt.legend()

plt.show()
```
